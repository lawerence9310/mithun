```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/hr_dataset.csv')
data.head()
```

```python
data.info()
```

```python
data = pd.get_dummies(data)
data
```

```python
data.info()
```

```python
data = pd.get_dummies(data,dtype=int)
data.head()
```

```python
x = data.drop(['churn'],axis=1)
y=data.churn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model = models.Sequential()

model.add(layers.Dense(8, input_shape =(20,),activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",metrics=['accuracy'],optimizer='adam')
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
model1 = models.Sequential()

model1.add(layers.Dense(16, input_shape =(20,),activation='relu'))

model.add(layers.Dense(16,activation='relu'))
model.add(layers.Dense(1,activation='relu'))

model1.add(layers.Dense(16,activation='sigmoid'))
```

```python
model1.add(layers.Dense(1,activation='sigmoid'))
```

```python
model1.summary()
```

```python
model1.compile(loss="binary_crossentropy",metrics=['accuracy'],optimizer='adam')
```

```python
model1.fit(x_train,y_train,epochs=20)
```

print("Accuray, loss of the model", model.evaluate(x_test,y_test))
print("Accuray, loss of the model", model1.evaluate(x_test,y_test)
